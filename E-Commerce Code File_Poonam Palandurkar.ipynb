```python
#Q5
def Ecom_revenue(discounted_price):

    if discounted_price>600:
        return discounted_price*(25/100)
    elif (discounted_price >350 and discounted_price <=600):
        return discounted_price*(15/100)
    elif (discounted_price >100 and discounted_price <=350):
        return discounted_price*(10/100)
    else:
        return discounted_price*(5/100)
```

```python
#Q5
ECom_Data['Total_revenue']=ECom_Data['discounted_price'].apply(Ecom_revenue)
ECom_Data.head(5)

#the total (sum) of all to get total Revenue.
totalrevenuesum=ECom_Data.Total_revenue.sum()
totalrevenuesum
```
